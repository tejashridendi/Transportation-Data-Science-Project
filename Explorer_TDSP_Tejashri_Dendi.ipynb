##<font color='crimson'> **Time Series Analysis**</font>
GOAL: The main goal is to dive deeper into Time Series Analysis in order to better understand our data's trends over time.

**Step 1:**

Time Series Analysis (TSA) - Trends, Seasonality, and Residual Components.

References:


*  [ Time Series patterns here](https://otexts.com/fpp2/tspatterns.html)
* [ Time Plots here](https://otexts.com/fpp2/time-plots.html)
*[Decomposition of Time Series Data into Trend and Seasonality](https://machinelearningmastery.com/decompose-time-series-data-trend-seasonality/)


**TSA:**
> *  Time series data records quantities that represent or trace the values taken by, a variable over a period such as a month, quarter, or year. In other words, time series data describes something over time.

> * The difficulty with time series is that it is not a binary task. If your test forecast is the same as your original data, there is a great great chance that your model is overfitting your data.

> * In time series analysis, trends, seasonality, and residual components are the three main parts that make up a time series dataset:

*Trend*

The long-term direction or movement of the data, which can be gradual, decreasing, or stable over time.

*Seasonality*

Short-term, repeating patterns that occur at fixed intervals, such as monthly, quarterly, or yearly. These patterns can be caused by factors like seasons or cycles, such as increased sales during the holidays.

*Residual*

Also known as error or irregular component, this is the random fluctuations or noise that remains after removing the trend and seasonality. It can also include any patterns that aren't captured by the other components.







**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
#file_path = "__________"
#data = pd.read_csv(file_path)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M')

# Time of Day Analysis
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the average number of crashes per hour
average_crashes_per_hour = data.groupby('Hour of Day').size() / data['Hour of Day'].nunique()

# Plot the average number of crashes
plt.figure(figsize=(6, 3))
sns.barplot(x=average_crashes_per_hour.index, y=average_crashes_per_hour.values)

plt.xlabel('Hour of Day')
plt.ylabel('Average Number of crashes')

plt.xticks(range(0, 24))
plt.title('Average Number of crashes per Hour of Day')
plt.show()


````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals.

Reference: [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared)  about trends, components, and further analysis!

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby(data['CRASH DATE']).size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Date')
plt.ylabel('Number of Motor Vehicle Collisions in NYC')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonality ')
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

**Step 3:** Next, we will begin "Severity Mapping." We'll plot crashes on the map and code them based on severity, distinguishing between crashes that resulted in injuries and those that led to fatalities. This will give us a visual representation of where severe crashes tend to occur. </br>

Let's make a map that's inclusive for people with color blindness by also creating differently-shaped markers (squares, circles, and triangles) for crashes, injuries, and fatalities.

````python
# Continue building your heatmap
# Sample a subset of the data for visualization
sample_data_severity = data_geo.sample(n=1000, random_state=42)

# Create a base map
m_severity = folium.Map(location=[40.730610, -73.935242], zoom_start=10)

# Add crashes to the map with color coding and shape coding based on severity
for index, row in sample_data_severity.iterrows():
    if row['NUMBER OF PERSONS KILLED'] > 0:
        color = "red"  # Fatalities

        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=3,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


    elif row['NUMBER OF PERSONS INJURED'] > 0:
        color = "yellow"  # Injuries
        folium.CircleMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          radius=5,
          color=color,
          fill=True,
          fill_color=color
       ).add_to(m_severity)
    else:
        color = "green"  # No injuries or fatalities
        folium.features.RegularPolygonMarker(
          location=[row['LATITUDE'], row['LONGITUDE']],
          number_of_sides=4,
          radius=5,
          gradient = False,
          color=color,
          fill=True,
          fill_color=color
        ).add_to(m_severity)


m_severity.save("severity.html")
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
# file_path = "__________"
# data = pd.read_csv(file_path)

# Convert 'CRASH DATE' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])

# Extract the day of the week (Monday=0, Sunday=6)
data['Day of Week'] = data['CRASH DATE'].dt.day_name()

# Group by 'Day of Week' and calculate the total number of crashes, injuries, and fatalities
crashes_by_day = data.groupby('Day of Week').size()
injuries_by_day = data.groupby('Day of Week')['NUMBER OF PERSONS INJURED'].sum()
fatalities_by_day = data.groupby('Day of Week')['NUMBER OF PERSONS KILLED'].sum()

# Create a dataframe for easy plotting
summary_by_day = pd.DataFrame({
    'Crashes': crashes_by_day,
    'Injuries': injuries_by_day,
    'Fatalities': fatalities_by_day
}).reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Plot the total number of crashes, injuries, and fatalities per day of the week
plt.figure(figsize=(9, 6))

# Plotting crashes
plt.subplot(3, 3, 1)
sns.barplot(x=summary_by_day.index, y=summary_by_day['Crashes'], palette='Blues_d')
#plt.title('Total Number of Crashes by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crashes')
plt.xticks(rotation=45, ha='right')
#plt.tight_layout()
#plt.show()

# Plotting injuries
plt.subplot(3, 3, 2)
sns.barplot(x=summary_by_day.index, y=summary_by_day['Injuries'], palette='Oranges_d')
#plt.title('Total Number of Injuries by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Injuries')
plt.xticks(rotation=45, ha='right')

# Plotting fatalities
plt.subplot(3, 3, 3)
#plt.figure(figsize=(6, 3))
sns.barplot(x=summary_by_day.index, y=summary_by_day['Fatalities'], palette='Reds_d')
#plt.title('Total Number of Fatalities by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Fatalities')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()




````

**The highest number of crashes, injuries and fatalities occurred on Friday, Friday, Saturday and the least occurred on Sunday, Sunday and Wednesday respectively.**

# Time of Day Analysis on Fridays:

````python
# Filter data for Fridays
friday_data = data[data['Day of Week'] == 'Friday']

# Extract hour of the day
friday_data['Hour of Day'] = friday_data['CRASH TIME'].dt.hour

# Group by hour and count crashes
crashes_by_hour_friday = friday_data.groupby('Hour of Day').size()

# Plot the distribution
plt.figure(figsize=(12, 6))

sns.barplot(x=crashes_by_hour_friday.index, y=crashes_by_hour_friday.values)
plt.title('Number of Crashes on Fridays by Hour of Day')
plt.xlabel('Hour of Friday')
plt.ylabel('No. of Crashes')


plt.tight_layout()
plt.show()
````

**Findings**:

Spikes are seen during rush hours notably 8 AM in the morning hours (7 AM -10 AM), 12 pm lunch hour and 5PM in the evening hours, it  suggests a correlation with increased traffic. Similarly, late-night spikes are seen and remained at same  range for a few hours starting from 8PM - 12PM until midnight when compared with early morning hours, could hint at potential alcohol-related incidents.

#  Exploring Contributing Factors on Fridays:
